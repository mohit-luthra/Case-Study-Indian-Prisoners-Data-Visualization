```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from matplotlib.ticker import FuncFormatter


data = pd.read_excel("Indian_Prisoners.xlsx")  
data.columns = ["Country", "Indian Prisoners"]
```

```python
data = data.sort_values(by="Indian Prisoners", ascending=False)
top_10 = data.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_10["Indian Prisoners"], 
    y=top_10["Country"], 
    palette="Blues_r")

for index, value in enumerate(top_10["Indian Prisoners"]):
    plt.text(value + 50, index, str(value), va='center', fontsize=12, fontweight='bold', color='black')

# Customization
plt.title("Top 10 Countries with Highest Number of Indians in Foreign Jails", fontsize=16, fontfamily="serif",fontweight='bold', pad=35)
plt.xlabel("Number of Indian Prisoners", fontsize=13)
plt.ylabel("Country", fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', alpha=0.2)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["left"].set_visible(False)

plt.figtext(0.9, -0.05, "Source: Ministry of External Affairs, India | Visualisation by Mohit Luthra", horizontalalignment='right', fontsize=10, color='gray')
plt.show()

```

```python
# Choropleth Map (World Heatmap)
fig = px.choropleth(
    data,
    locations="Country",
    locationmode="country names",
    color="Indian Prisoners",
    hover_name="Country",
    color_continuous_scale="reds",
    title="🌍 Where are Indians Imprisoned Abroad? (2024)")

# Customize layout
fig.update_layout(
    title_font=dict(size=18, family="Arial", color="black"),
    geo=dict(showframe=False, showcoastlines=True, projection_type="natural earth"),
    coloraxis_colorbar=dict(title="Number of Indian Prisoners"),
)

# Show plot
fig.show()
```

```python
data = pd.read_excel("death_sentence.xlsx") 
data.columns = ["Country", "Number of Prisoners"]
```

```python
data = data.sort_values(by="Number of Prisoners", ascending=False)
top_10 = data.head(10)

import matplotlib.pyplot as plt
import seaborn as sns

top_10["Country"] = top_10["Country"].str.replace(" ", "\n")  

plt.figure(figsize=(8, 6))

sns.barplot(
    x=top_10["Country"], 
    y=top_10["Number of Prisoners"], 
    palette="YlOrBr_r"
)

for index, value in enumerate(top_10["Number of Prisoners"]):
    plt.text(index, value + 0.4, str(value), ha='center', fontsize=12, fontweight='bold', color='black')

plt.title("Countries with Highest Number of Indian Nationals on Death Row", fontsize=16, fontfamily="serif",fontweight='bold', pad=20)
plt.xlabel("Country", fontsize=13)
plt.ylabel("Number of Indian Prisoners", fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', alpha=0.4)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["left"].set_visible(False)

plt.figtext(0.9, -0.1, "Source: Ministry of External Affairs, India | Visualisation by Mohit Luthra", horizontalalignment='right', fontsize=10, color='gray')
plt.show()
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt

data = dict(zip(top_10["Country"], top_10["Number of Prisoners"]))

fig = plt.figure(FigureClass=Waffle, rows=10, values=data, 
                 title={"label": "Indian Nationals on Death Row", "loc": "center"},
                 legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': 2})
plt.show()
```
